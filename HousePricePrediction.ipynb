```python
## Loading Dataset to fit in the Regression Model

import pandas as pd
import numpy as nm
from sklearn.preprocessing import MinMaxScaler

data = pd.read_csv("https://raw.githubusercontent.com/itsanikesh03/ME698/main/housing.csv",header = None)


## Scaling data
scaler = MinMaxScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(data))

X = scaled_data.iloc[:,:-1]
Y = scaled_data.iloc[:,13]

#print(X.shape,Y.shape,data.shape,)
```

```python
data.head()
```

```python
## Splitting data into training and testing data

from sklearn.model_selection import train_test_split

xtr,xt,ytr,yt = train_test_split(X,Y,test_size = 0.1,random_state=1)
```

```python
## Adding a column to calculate Bias

x_train = nm.c_[nm.ones(len(xtr)),xtr]
x_test = nm.c_[nm.ones(len(xt)),xt]
```

```python
## Training Of Model

W = nm.ones(14)
W_grad = nm.zeros(14)
iteration = 20000
learn_rate = 0.001

for i in range(iteration):
    yp = x_train@W
    n = len(x_train)
    for j in range(len(W)):
        W_grad[j%14] = -(2/n)*sum((x_train[:,j%14])*(ytr-yp))
        W[j%14] = W[j%14] - learn_rate*W_grad[j%14]
    
    MSE = (sum((ytr-yp)**2))/n 
    #print("iter {}, MSE {}".format(i, MSE))
  



```

```python
## Predicted Price for given Test set
y_pred = x_test@W

# Scaled test dataset with scaled Price in last column
table=pd.DataFrame(nm.c_[xt,y_pred])

# Table with original test sample containing predicted price in last column
TABLE = pd.DataFrame(scaler.inverse_transform(table))
TABLE.columns = ['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','X11','X12','X13','Predicted_Y']
# TABLE.head()

Pred_Price = TABLE['Predicted_Y']
```

```python
# Scaled test table with Target Prices
DATA = pd.DataFrame(scaler.inverse_transform(pd.DataFrame(nm.c_[xt,yt])))
DATA.columns = ['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','X11','X12','X13','Y']
#  DATA.head()

## Mean Absolute Error
k = len(yt)
MAE = sum(abs(DATA['Y']-TABLE['Predicted_Y']))/k
```

```python
print("Weights:",W)
print("Predicted Value for test data:",Pred_Price)
print("Mean Absolute Error:",MAE)
```
